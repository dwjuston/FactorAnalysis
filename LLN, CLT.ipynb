```python
arr=np.random.exponential(scale=0.5, size=100)
```

#### mean = 1/lambda, sigma = 1/lambda -> mean=0.5, lambda=2, sigma=0.5

```python
trials=[]
for i in range(50):
    trials.append(np.random.exponential(scale=0.5, size=100))
```

```python
pd.DataFrame(trials)
```

```python
pd.DataFrame(trials).mean(axis=1).plot.hist()
```

#### 1 power law is $f(x) = a x^{-b} + c$

```python
def power_law(a, b, c):
    def f(x):
        return a * np.power(x, -b) + c
    return f
```

```python
x_arr = np.linspace(0.5, 3, 1000)
```

```python
pd.DataFrame({
    'y, b=2':power_law(1,2,0)(x_arr),
    'y, b=3':power_law(1,3,0)(x_arr)
},index=x_arr).plot()
```

#### 1-2 log-log plot

```python
pd.DataFrame({
    'y, b=2':np.log(power_law(1,2,0)(x_arr)),
    'y, b=3':np.log(power_law(1,3,0)(x_arr))
},index=np.log(x_arr)).plot()
```

#### 2 decay

```python
def exponential_decay(a, b, c):
    def f(x):
        return a * np.exp(-b*x) + c
    return f
```

```python
def exponential_squared_decay(a, b, c):
    def f(x):
        return a * np.exp(-b*np.square(x)) + c
    return f
```

```python
x_arr = np.linspace(0.5, 4, 1000)
pd.DataFrame({
    'exponential decay': exponential_decay(1,1,0)(x_arr),
    'power decay': power_law(1,2,0)(x_arr),
    'exponential squared decay': exponential_squared_decay(1,1,0)(x_arr),
},index=x_arr).plot()
```

```python
x_arr = np.linspace(1, 5, 1000)
pd.DataFrame({
    'exponential decay': exponential_decay(1,1,0)(x_arr),
    'power decay': power_law(1,2,0)(x_arr),
    'exponential squared decay': exponential_squared_decay(1,1,0)(x_arr),
},index=x_arr).plot()
```

```python
x_arr = np.linspace(2, 10, 1000)
pd.DataFrame({
    'exponential decay': exponential_decay(1,1,0)(x_arr),
    'power decay': power_law(1,2,0)(x_arr),
    'exponential squared decay': exponential_squared_decay(1,1,0)(x_arr),
},index=x_arr).plot()
```

#### 3-1 upper tail of lognormal

```python
from scipy.stats import norm, lognorm
```

```python
x_arr = np.linspace(0.01, 10, 1000)

pd.DataFrame({
    'lognormal': lognorm.pdf(x_arr, 1),
},index=x_arr).plot()
```

```python
x_arr = np.linspace(0.5, 3, 1000)
pd.DataFrame({
    'lognormal': np.log(lognorm.pdf(x_arr, 1)),
},index=np.log(x_arr)).plot()
```

```python
x_arr = np.linspace(3, 20, 1000)
pd.DataFrame({
    'lognormal': np.log(lognorm.pdf(x_arr, 1)),
},index=np.log(x_arr)).plot()
```

```python
x_arr = np.linspace(10, 20, 1000)
pd.DataFrame({
    'lognormal': np.log(lognorm.pdf(x_arr, 1)),
},index=np.log(x_arr)).plot()
```

#### 3-2 t-distribution

```python
from scipy.stats import t
```

```python
x_arr = np.linspace(-3, 3, 1000)
pd.DataFrame({
    't_distribution': t.pdf(x_arr, df=5),
    'normal': norm.pdf(x_arr)
},index=x_arr).plot()
```

```python
x_arr = np.linspace(5, 20, 1000)
pd.DataFrame({
    't_distribution log-log': np.log(t.pdf(x_arr, df=5)),
},index=np.log(x_arr)).plot()
```

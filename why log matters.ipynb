```python
simple_list = list(range(1,11))
```

#### At first glance, it is a very good list of numbers

```python
simple_list
```

#### But indeed, its growth rate is decaying

```python
import pandas as pd
import numpy as np
```

```python
pd.Series(simple_list).pct_change().plot()
```

#### Take log can change the way we view this list: it is growing more and more slowly

```python
pd.Series(np.log(simple_list)).plot()
```

```python
pd.Series(np.log(simple_list)).diff().plot()
```

#### After taking the log, the perfectly natural list of numbers is exponential growth

```python
pd.Series(np.exp(simple_list)).plot()
```

```python
pd.Series(np.log(np.exp(simple_list))).plot()
```

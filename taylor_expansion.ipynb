```python
import numpy as np
import pandas as pd
from numpy import exp, sin, cos
```

```python
def f(x):
    return exp(sin(x))
```

```python
f(0)
```

```python
f(0.5)
```

```python
def f1(x):
    return exp(sin(x)) * cos(x)

def f2(x):
    return exp(sin(x)) * cos(x)**2 - exp(sin(x)) * sin(x)
```

#### 一阶泰勒展开

```python
a = 0
x = 0.5
f(a) + f1(a) * (x-a)
```

```python
a = 0
x = 0.5
f(a) + f1(a) * (x-a) + 0.5 * f2(a) * (x-a)**2
```

```python
arr=np.arange(-2,2,0.05)
```

```python
a = 0
true = [f(x) for x in arr]
taylor1 = [f(a) + f1(a) * (x-a) for x in arr]
taylor2 = [f(a) + f1(a) * (x-a) + 0.5 * f2(a) * (x-a)**2 for x in arr]
```

```python
pd.DataFrame({"true":true, "taylor1":taylor1, "taylor2":taylor2}, index=arr).plot(figsize=(10,10))
```

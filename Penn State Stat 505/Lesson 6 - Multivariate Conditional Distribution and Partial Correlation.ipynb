#### Contidional Distribution Properties

The conditional distribution of $X_1$ given known values for $X_2 = x_2$ is a multivariate normal with:
$$
mean = \mu_1 + \Sigma_{12} \Sigma_{22}^{-1} (x_2 - \mu_2)
$$

$$
cov mat = \Sigma_{11} - \Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21}
$$
